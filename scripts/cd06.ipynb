# Training and Visualizing a Decision Tree

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
from IPython.display import Image
```

```python
iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
export_graphviz(
    tree_clf,
    out_file="iris_tree.dot",
    feature_names=iris.feature_names[2:],
    class_names=iris.target_names,
    rounded=True,
    filled=True)
```

```python
Image(filename='iris_tree.png')
```

# Estimating Class Probabilities

```python
tree_clf.predict_proba([[5, 1.5]])
```

```python
tree_clf.predict([[5, 1.5]])
```

# Regression

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X = 6 * np.random.rand(500, 1) - 3
```

```python
y = 2 + X**2 + 0.50*np.random.randn(500, 1)
```

```python
plt.plot(X, y, "b.")
plt.axis([-3, 3, 0, 13])
```

```python
tree_reg = DecisionTreeRegressor(max_depth=3, min_samples_leaf=20)
tree_reg.fit(X, y)
```

```python
export_graphviz(
    tree_reg,
    out_file="quadratic.dot",
    feature_names="x",
    class_names="y",
    rounded=True,
    filled=True)
```

```python
Image(filename='quadratic.png')
```
